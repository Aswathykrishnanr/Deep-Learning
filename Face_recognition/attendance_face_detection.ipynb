```python
def face_encodings(images):

    encode_list=[]


    for i in images:
        img=cv2.cvtColor(i,cv2.COLOR_BGR2RGB)
        face_in_frame=fr.face_locations(img)
        encoder=fr.face_encodings(img,face_in_frame)[0]
        encode_list.append(encoder)
    return encode_list
```

```python
encode_known_faces=face_encodings(imgs)
```

```python
from datetime import datetime
```

```python
video=cv2.VideoCapture(0)

while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    face_in_frame=fr.face_locations(img1)
    encoded=fr.face_encodings(img,face_in_frame)


    today = datetime.now().strftime("%Y-%m-%d")
    file_name = f"attendance_{today}.csv"

    def mark_attendance(name):
        with open(file_name, "a") as f:
            time_now = datetime.now().strftime("%H:%M:%S")
            f.write(f"{name},{time_now}\n")



    for face_loc,enc in zip(face_in_frame,encoded):
        matches=fr.compare_faces(encode_known_faces,enc)
        face_distance=fr.face_distance(encode_known_faces,enc)

        match_index=np.argmin(face_distance)
        if matches[match_index]==True:
            name=class_names[match_index]
            mark_attendance(name)
            

        else:
            pass
            
        


    cv2.imshow('video',img)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break

video.release()
cv2.destroyAllWindows()
            
```
