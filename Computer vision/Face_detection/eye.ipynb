```python
import cv2
import matplotlib.pyplot as plt
```

```python
img1=cv2.imread("Eyes.jpg")
conv_img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img1)
```

```python
eyes_data=cv2.CascadeClassifier("haarcascade_eye (1).xml")
eyes_data
```

```python
print(eyes_data.empty())
```

```python
#gray = cv2.cvtColor(conv_img1, cv2.COLOR_BGR2GRAY) 
eyes=eyes_data.detectMultiScale(conv_img1,minNeighbors=5)
eyes
```

```python
for (x,y,w,h) in eyes:
    cv2.rectangle(conv_img1,(x,y),(x+w,y+h),color=(255,0,0),thickness=1)
```

```python
plt.imshow(conv_img1)
```

```python
video = cv2.VideoCapture('vdo.mp4')

while True:
    succ,img = video.read()
    if succ:
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

        eye = eyes_data.detectMultiScale(gray,minNeighbors=17)

        for (x,y,w,h) in eye:
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
        
        cv2.imshow('Video',img)

        if cv2.waitKey(1) & 0XFF == ord('q'):
            break
    else:
        print("Video Ended")
        break

video.release()
cv2.destroyAllWindows()
```
