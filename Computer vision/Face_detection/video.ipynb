```python
import cv2
```

```python
face_cas=cv2.CascadeClassifier('haarcascade_frontalface_default (1).xml')
face_cas
```

```python
#video
video=cv2.VideoCapture('vdo.mp4')                                    #0 for using own camera #1 for other webcams
#read()
while True:                                                          #read 2 variable sucess,img
    sucess,img=video.read()                                          #BGR format 
    if sucess:
        gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)                #sucess
        faces=face_cas.detectMultiScale(gray_img,minNeighbors=5)     #gray_image format
        for(x,y,w,h) in faces:
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),4)           #img is in BGR
        cv2.imshow('video',img)
        if cv2.waitKey(1) & 0XFF==ord('q'):                          #to close the window using key 'q'
            break
    else:
        print("Finished")
        break
video.release()                                                      #video closes
cv2.destroyAllWindows()                                              #closes all windows
```
