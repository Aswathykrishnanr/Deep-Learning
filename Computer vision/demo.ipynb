```python
#convert image into gray scale
img_gray=cv2.imread('baby_girl.jpg',cv2.IMREAD_GRAYSCALE)
img_gray
```

```python
img_gray.shape
```

```python
cv2.imshow('Baby image',img_gray)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
#save image 
cv2.imwrite('Baby_girl_gray.jpg',img_gray)
```

```python
img_gray2=cv2.imread('baby_girl2.jpg',cv2.IMREAD_GRAYSCALE)
img_gray2
```

```python
img_gray2.shape
```

```python
cv2.imshow("Baby_image_gray",img_gray2)
cv2.waitKey(5000)
cv2.destroyAllWindows
```

```python
cv2.imwrite('Baby_girl_gray2.jpg',img_gray2)
```

```python
#convert image to binary 
#threshold
#2 variable
thres,bin_img=cv2.threshold(img_gray,100,255,cv2.THRESH_BINARY)           
cv2.imshow('binary_image',bin_img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
thres
```

```python
cv2.imwrite('Baby_girl_binary.jpg',bin_img)
```

```python
#THRESH_OTSU  to assign  same ratio of color by machine
thres1,bin_img1=cv2.threshold(img_gray,100,255,cv2.THRESH_OTSU)           
cv2.imshow('binary_image',bin_img1)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
thres1
```

```python
thres2,bin_img2=cv2.threshold(img_gray2,150,255,cv2.THRESH_BINARY)         
cv2.imshow('binary_image 2',bin_img2)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```
