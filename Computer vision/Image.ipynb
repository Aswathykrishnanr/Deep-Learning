```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('baby_girl.jpg')
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
#Bounding Box
#Rectangle
#Left Top,Right Bottom
cv2.rectangle(img1,pt1=(90,40),pt2=(200,150),color=(255,0,0),thickness=2)
plt.imshow(img1)
```

```python
img2=cv2.imread('baby_girl2.jpg')
img21=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.imshow(img21)
```

```python
cv2.rectangle(img21,pt1=(40,50),pt2=(180,150),color=(255,0,0),thickness=2)
plt.imshow(img21)
```

```python
#circle
img3=cv2.imread("Baby_boy2.jpg")
img31=cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
plt.imshow(img31)
```

```python
cv2.circle(img31,center=(100,100),radius=60,color=(0,0,255),thickness=2)
plt.imshow(img31)
```

```python
#text
img4=cv2.imread("Baby_boy.jpg")
img41=cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)
plt.imshow(img41)
```

```python
#putText
cv2.putText(img41,text='Cute baby',org=(10,50),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=1,color=(0,255,0),thickness=1)
plt.imshow(img41)
```
