```python
#hand landmarks
#pip install mediapipe
import mediapipe as mp
import cv2
```

```python
#hand landmark detection
mphands=mp.solutions.hands   #hand detection module
mp_drawing=mp.solutions.drawing_utils  #hands-sub module
```

```python
#hand submodule- class name hands()
data=mphands.Hands()
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result=data.process(img1)
    print(result.multi_hand_landmarks)
    if result.multi_hand_landmarks:
        for x in result.multi_hand_landmarks:
            mp_drawing.draw_landmarks(img,x,mphands.HAND_CONNECTIONS,mp_drawing.DrawingSpec(color=(255,0,0),thickness=5,circle_radius=3),mp_drawing.DrawingSpec(color=(0,0,255),thickness=5))


    cv2.imshow('Hands',img)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break

video.release()
cv2.destroyAllWindows()
```
