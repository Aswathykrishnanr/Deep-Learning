```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('face_detection1.jpg')
plt.imshow(img)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
face_data=cv2.CascadeClassifier('haarcascade_frontalface_default (1).xml')
face_data
```

```python
#Face detect
faces=face_data.detectMultiScale(conv_img,minNeighbors=5)
faces                                            #array[[left top 2values,height,width]] of 2 faces
```

```python
for(x,y,w,h) in faces:
    cv2.rectangle(conv_img,pt1=(x,y),pt2=(x+w,y+h),color=(255,0,0),thickness=2)
```

```python
plt.imshow(conv_img)
```

```python
img2=cv2.imread('Stranger-Things.jpg')
conv_img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img2)
```

```python
faces2=face_data.detectMultiScale(conv_img2,minNeighbors=5)
faces2
```

```python
for(x,y,w,h) in faces2:
    cv2.rectangle(conv_img2,pt1=(x,y),pt2=(x+w,y+h),color=(255,0,0),thickness=3)
```
