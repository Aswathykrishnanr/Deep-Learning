```python
import cv2
import pytesseract
```

```python
img=cv2.imread("plate_2.jpg")
cv2.imshow("Number_plate",img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
pytesseract.pytesseract.tesseract_cmd=r'C:\Program Files\Tesseract-OCR\tesseract.exe'
```

```python
text=pytesseract.image_to_string(img)
print(text)
```

```python
import re
```

```python
plate_number = re.sub('[^A-Z0-9]', '', text)
```

```python
print("Detected Number Plate:", plate_number)
```
