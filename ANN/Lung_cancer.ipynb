```python
df.drop(['Name','Surname'],axis=1,inplace=True)
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=1)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#Model Creation
from keras.models import Sequential
from keras.layers import Dense
```

```python
ann=Sequential()  #initialisation
#input layer
ann.add(Dense(units=4,activation='relu'))   #units=no of artificial neurons
#Hidden layer
ann.add(Dense(units=10,activation='relu'))
#Hidden layer
ann.add(Dense(units=6,activation='relu'))
#Output layer
ann.add(Dense(units=1,activation='sigmoid'))   #binary classification-sigmoid
```

```python
#compile
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy']) #binary classification=binary cross entropy
ann.fit(x_train,y_train,epochs=100,batch_size=12)
```

```python
ann.evaluate(x_test,y_test)
```

```python
ann.summary()
```

```python
y_pred=ann.predict(scaler.transform([[35,3,5,4]]))
y_pred
if y_pred>0.5:
    print("Lung cancer") #1
else:
    print("Not a cancer") #0
```
